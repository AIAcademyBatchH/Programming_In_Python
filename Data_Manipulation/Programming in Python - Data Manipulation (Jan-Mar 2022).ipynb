```python
# groupby and aggregate
```

```python
# what is the average quantity sold by region
```

```python
data.groupby('Region',as_index=False).agg({'Quantity':np.mean})
```

```python
# what is the total sales by categories?
```

```python
data.groupby('Category',as_index=False).agg({'Sales':np.sum})
```

```python
# multi-level column indexing
```

```python
data.groupby('Category')[['Sales','Profit']].agg(['max','sum'])
```

```python
data.groupby('Category')[['Sales','Profit']].agg({'Sales':sum, 'Profit':max})
```

```python
data.query("Region=='East'").groupby('Category', as_index=False)[['Sales','Profit']].agg({'Sales':sum, 'Profit':max})
```

```python
# what is the average discount by segment?
```

```python
data.groupby('Segment', as_index=False).agg({'Discount': 'mean'})
```

```python
data.groupby('Segment', as_index=False).agg({'Discount': 'mean'}).rename(columns={'Discount':'Average Discount'})
```

```python
data.groupby('Segment', as_index=False).agg({'Discount': 'mean'}).rename(columns={'Discount':'Average Discount'}).sort_values('Average Discount',ascending=False)
```

```python
# which segment of customers are most profitable?
```

```python
data.groupby('Segment', as_index=False).agg({'Profit': 'sum'}).rename(columns = {'Profit': 'Profit_Sum'}).\
sort_values('Profit_Sum', ascending=False)
```

```python
data.groupby('Segment', as_index=False).agg({'Profit': 'mean'}).rename(columns = {'Profit': 'Profit_Avg'}).\
sort_values('Profit_Avg', ascending=False)
```

```python
# Top 5 States by total profit?
```

```python
data.groupby('State', as_index=False).agg({'Profit': 'sum'}).rename(columns = {'Profit': 'Profit_Sum'}).\
sort_values('Profit_Sum', ascending=False).head(5)
```
